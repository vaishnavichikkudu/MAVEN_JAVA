```python
from datasets import load_dataset
from transformers import LlamaForCausalLM, AutoTokenizer, Trainer, TrainingArguments
from peft import get_peft_model, LoraConfig, TaskType
```

```python
# Load the dataset split (train)
dataset = load_dataset('VishaalY/synthetic-code-generations', split='train')
```

```python
# Split the dataset into train and evaluation sets
eval_dataset = dataset.select(range(1000))  # Use the first 1000 samples for evaluation
train_dataset = dataset.select(range(1000, len(dataset)))  # Use the rest for training
```

```python
# Load the pre-trained LLaMA 3 model and tokenizer
model_name = "meta-llama/Llama-3.2-1B"
model = LlamaForCausalLM.from_pretrained(
    model_name,
    device_map="auto",  # Automatically offload parts of the model to CPU if needed
    torch_dtype="float32",  # Explicitly set to float32 to avoid FP16
    use_cache=False  # Disable cache to work with gradient checkpointing
)
```

```python
# Enable gradient checkpointing after loading the model
model.gradient_checkpointing_enable()
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Set a pad_token for the tokenizer if not already set (to handle padding properly)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token if tokenizer.eos_token else "[PAD]"
    if tokenizer.pad_token == "[PAD]":
        tokenizer.add_special_tokens({'pad_token': '[PAD]'})
```

```python
# Tokenize the dataset and set labels for code generation
def tokenize_function(examples):
    encoding = tokenizer(examples['response'], padding="max_length", truncation=True, max_length=400)
    encoding['labels'] = encoding['input_ids']  # Set labels equal to input_ids for causal LM
    return encoding
```

```python
# Tokenize the train and eval datasets
tokenized_train_dataset = train_dataset.map(tokenize_function, batched=True)
tokenized_eval_dataset = eval_dataset.map(tokenize_function, batched=True)
```

```python
# Configure LoRA
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,  # Type of task
    inference_mode=False,          # Set to False for training
    r=8,                           # LoRA rank (number of trainable parameters per layer)
    lora_alpha=16,                 # Scaling factor for LoRA
    lora_dropout=0.1               # Dropout rate for LoRA
)
```

```python
# Wrap the model with PEFT
model = get_peft_model(model, lora_config)
```

```python
# Print the trainable parameters for verification
model.print_trainable_parameters()
```

```python
# Calculate the number of training steps based on dataset size
train_size = len(tokenized_train_dataset)  # Total number of training samples
batch_size = 2  # Reduced batch size for lower GPU usage
num_train_steps = train_size // batch_size
```

```python
# Set up the training arguments with optimizations for memory usage
training_args = TrainingArguments(
    output_dir="./results",            # Output directory
    #evaluation_strategy="epoch",       # Evaluation strategy
    learning_rate=2e-4,                # Learning rate
    per_device_train_batch_size=batch_size,  # Reduced batch size for training
    per_device_eval_batch_size=batch_size,   # Batch size for evaluation
    num_train_epochs=3,                # Number of training epochs
    weight_decay=0.01,                 # Weight decay
    logging_dir="./logs",              # Directory for logs
    logging_steps=10,                  # Log every 10 steps
    save_steps=500,                    # Save checkpoint every 500 steps
    save_total_limit=2,                # Limit number of saved checkpoints
    max_steps=3000,         # Manually specify the max number of training steps
    report_to=["none"],                # Disable reporting to external services
    gradient_accumulation_steps=4,     # Accumulate gradients to simulate a larger batch size
    fp16=False,                         # Disable FP16
    dataloader_num_workers=2,          # Number of workers for data loading
    group_by_length=True,
    warmup_ratio = 0.03,
    warmup_steps = 5,
    #save_steps = 0,
    lr_scheduler_type="linear",
)
```

```python
# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train_dataset,  # Use the tokenized train dataset
    eval_dataset=tokenized_eval_dataset,    # Use the tokenized eval dataset
)
```

```python
# Start training
trainer.train()
```
